```python
# https://www.pyimagesearch.com/2016/07/25/convolutions-with-opencv-and-python/

%pylab inline
import cv2

ball = cv2.imread('img/ball.png')
ball_gray = cv2.cvtColor(ball, cv2.COLOR_BGR2GRAY)

kernel = np.array([[-1, 1], [-1, 1]])
size = kernel.shape[0]

margin = size - 1
pixels = []

for row_index, row in enumerate(ball_gray[:-margin]):
    for column_index, pixel in enumerate(row[:-margin]):
        image_part = ball_gray[row_index:row_index + size, column_index:column_index + size]
        pixels.append((kernel * image_part).sum())
```

```python
ball_out = np.array(pixels).reshape(235, 318)
plt.imshow(ball_out)
```
